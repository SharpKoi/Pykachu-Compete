```python
import pandas as pd
import numpy as np
import collections
```

```python
df = pd.read_csv("order_brush_order.csv");
df
```

```python
time_zero = pd.to_datetime('2019-12-27')
df['unix_time'] = pd.to_datetime(df['event_time']).apply(lambda x: int((x-time_zero).total_seconds()))
df.sort_values(['shopid', 'unix_time'])
```

```python
shops = collections.defaultdict(list)
orders = df.drop(['orderid', 'event_time'], axis = 1).values
for shopid, userid, unix_time in orders:
    shops[shopid].append((userid, unix_time))

shops
```
