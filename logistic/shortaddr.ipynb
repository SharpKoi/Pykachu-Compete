```python
addr_list = ['Metro Manila', 'Luzon', 'Visayas', 'Mindanao']
SLA = pd.DataFrame([[3, 5, 7, 7], [5, 5, 7, 7], [7, 7, 7, 7], [7, 7, 7, 7]], 
                   index = addr_list,
                   columns = addr_list)
SLA
```

```python
dodf = pd.read_csv('delivery_orders_march.csv')
```

```python
# 看看資料長怎樣
dodf.head()
```

```python
total_addr = dodf['buyeraddress'].count()

# 先看看前100個結果怎麼樣
for i in range(100):
    addr = dodf['buyeraddress'].iloc[i]
    if 'Metro' in addr:
        dodf['buyeraddress'].iloc[i] = 'Metro Manila'
    elif 'Luzon' in addr:
        dodf['buyeraddress'].iloc[i] = 'Luzon'
    elif 'Visayas' in addr:
        dodf['buyeraddress'].iloc[i] = 'Visayas'
    elif 'Mindanao' in addr:
        dodf['buyeraddress'].iloc[i] = 'Mindanao'

dodf[:100]
```

```python
# 有些沒設定到 查看其中一列的address發生了什麼事
print(dodf.iloc[2]['buyeraddress'])
# 小寫也要考慮進去
```

```python
# 改一下前面的方法再做一次 這次做1000條
for i in range(1000):
    addr = dodf['buyeraddress'].iloc[i]
    if ('Metro' in addr or 'metro' in addr):
        dodf['buyeraddress'].iloc[i] = 'Metro Manila'
    elif ('Luzon' in addr or 'luzon' in addr):
        dodf['buyeraddress'].iloc[i] = 'Luzon'
    elif ('Visayas' in addr or 'visayas' in addr):
        dodf['buyeraddress'].iloc[i] = 'Visayas'
    elif ('Mindanao' in addr or 'mindanao' in addr):
        dodf['buyeraddress'].iloc[i] = 'Mindanao'

dodf[:1000]
# 
```

```python
# 用更有效率的方法(用temp當實驗品)
import re
temp = dodf

for addr in addr_list:
    addr_filter = temp.buyeraddress.str.contains(addr, flags=re.IGNORECASE)
    temp.buyeraddress[addr_filter] = addr
```

```python
temp
# 成功!
```

```python
#接下來把selleraddress也轉址
for addr in addr_list:
    addr_filter = temp.selleraddress.str.contains(addr, flags=re.IGNORECASE)
    temp.selleraddress[addr_filter] = addr
```
